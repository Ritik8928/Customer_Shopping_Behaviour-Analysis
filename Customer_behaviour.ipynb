```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.isnull().sum() #check missing values
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) #handle missing values
```

```python
df.isnull().sum().sum() #verify missing values
```

```python
df.columns = df.columns.str.lower()  #each word in snake casing or lower
df.columns = df.columns.str.replace(' ','_')  #remove the spaces between columns name
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
age_group = ['Young_adult','Adult','Middle_aged','Senior_citizen']
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_group)
```

```python
df[['age_group','age']].sample(5)
```

```python
#create column purchase frequency days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
    
}
df['Purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['Purchase_frequency_days','frequency_of_purchases']].head(10)
```
